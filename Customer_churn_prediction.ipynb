```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Churn_Modelling.csv")
```

```python
df.head()
```

```python
x = df.iloc[:,3: -1]
y = df.iloc[:, -1].values
```

```python
x.head()
```

```python
unique, counts = np.unique(y, return_counts=True)
print(np.asarray((unique, counts)))
```

```python
print(x['Geography'].value_counts())
print(x['Gender'].value_counts())
print(x['NumOfProducts'].value_counts())
print(x['HasCrCard'].value_counts())
print(x['IsActiveMember'].value_counts())
```

```python
x.describe()
```

```python
x.columns
```

```python
x.isnull().sum()
```

```python
x.dtypes
```

```python
x= pd.get_dummies(x, columns=['Geography', 'Gender'], drop_first=True)
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.15, stratify=y, random_state=0)
```

```python
x_train
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100)
rf.fit(x_train, y_train)
```

```python
y_pred = rf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
print(accuracy_score(y_test, y_pred))
```
